#### Using a list to contain another list:

```python
list_of_lists = [[3, 6], 1, 2, 'hello']
display(list_of_lists)
print(type(list_of_lists))
```

#### Using a tuple to contain a list:

```python
tuple_with_list = (1, [3, 6], 'world')
display(tuple_with_list)
print(type(tuple_with_list))
```

Keep in mind that while you can store a list inside a list or a tuple, you still cannot store a list directly inside a set because of the 'unhashable' nature of lists.
